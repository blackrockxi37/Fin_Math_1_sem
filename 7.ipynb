I: Задание 125   ТЗ 7-5
S: Для потока платежей  〖CF〗_1= {(1;200),   (2;300),   (3;200)}   приведённая величина при сложной процентной ставке 10% с точностью до 0,01 равна …  


```python
cf = [200, 300, 200]
periods = [1, 2, 3]
discount_rate = 0.1
pv = sum([cf[i] / (1 + discount_rate)**periods[i] for i in range(len(periods))])
print(f"Present Value: {pv:.2f}")
```

I: Задание 129   ТЗ 7-9
