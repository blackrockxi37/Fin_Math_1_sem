I: Задание 161   ТЗ 9-1
S: Семья планирует через 5 лет купить квартиру за 2500000 руб. и с этой целью ежемесячно на банковский депозит вносится определенная сумма. Годовая банковская ставка составляет 10% с ежемесячным начислением процентов. Ежемесячно вносимая сумма с точностью до 0,01 руб. равна …


```python
FV = 2500000
n = 5 * 12
i = 0.10 / 12

payment = FV / (((1 + i)**n - 1) / (i * (1 + i)**n))

print(f"Ежемесячный платёж: {payment:.2f}")
```
